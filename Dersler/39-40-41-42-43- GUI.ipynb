```python
#1. PART
#pack
#place
#grid
#add8e6
#Verdana 12 bold
from tkinter import *
from tkcalendar import DateEntry
master = Tk()
canvas = Canvas(master, height = 450, width = 750)
canvas.pack()

frame_ust = Frame(master, bg = "#add8e6")
frame_ust.place(relx=0.1,rely=0.1,relheight=0.1,relwidth=0.8)

frame_sol = Frame(master, bg = "#add8e6")
frame_sol.place(relx=0.1,rely=0.21,relheight=0.68,relwidth=0.3)

frame_sag_ana = Frame(master, bg = "#add8e6")
frame_sag_ana.place(relx=0.408,rely=0.21,relheight=0.68,relwidth=0.492)

hatirlatma_tipi_etiket = Label(frame_ust,bg="#add8e6",text="Hatırlatma Tipi:",font = "Verdana 12 bold")
hatirlatma_tipi_etiket.pack(padx=10,pady=10,side=LEFT)
hatirlatma_tipi_opsiyon = StringVar(frame_ust)
hatirlatma_tipi_acilir_menu = OptionMenu(frame_ust,
                                        hatirlatma_tipi_opsiyon,
                                        "Doğum günü",
                                        "Alışveriş",
                                        "Ödeme")
hatirlatma_tipi_acilir_menu.pack(padx=10,pady=10,side=LEFT)

hatirlatma_tarih_secici = DateEntry(frame_ust, width=12, background="orange",foreground="black",borderwidth=1,locale="tr_TR")
hatirlatma_tarih_secici._top_cal.overrideredirect(False)
hatirlatma_tarih_secici.pack(padx=10,pady=10,side=RIGHT)
hatirlatma_tarih_secici_etiketi = Label(frame_ust,bg="#add8e6",text="Hatırlatma Tarihi:",font="Verdana 12 bold")
hatirlatma_tarih_secici_etiketi.pack(padx=10,pady=10,side=RIGHT)


#2. PART
Label(frame_sol, text="Hatırlatma Yöntemleri:",font="Verdana 12 bold",bg="#add8e6").pack(padx=10,pady=10,anchor=NW)
var = IntVar()
R1 = Radiobutton(frame_sol,text="Sisteme Kaydet",variable = var,value=1,font="Verdana 10",bg="#add8e6")
R1.pack(anchor=NW,padx=10,pady=5)
R2 = Radiobutton(frame_sol,text="E-Posta Gönder",variable = var,value=2,font="Verdana 10",bg="#add8e6")
R2.pack(anchor=NW,padx=10,pady=5)
var1 = IntVar()
CB1 = Checkbutton(frame_sol,text="Bir hafta önce",font="Verdana 10",bg="#add8e6",variable=var1,onvalue=1,offvalue=0)
CB1.pack(anchor=NW,padx=25,pady=3)
var2 = IntVar()
CB2 = Checkbutton(frame_sol,text="Bir gün önce",font="Verdana 10",bg="#add8e6",variable=var2,onvalue=1,offvalue=0)
CB2.pack(anchor=NW,padx=25,pady=3)
var3 = IntVar()
CB3 = Checkbutton(frame_sol,text="Aynı gün",font="Verdana 10",bg="#add8e6",variable=var3,onvalue=1,offvalue=0)
CB3.pack(anchor=NW,padx=25,pady=3)


#3. PART
Label(frame_sag_ana,text="Hatırlatma Mesajı:",font="Verdana 12 bold",bg="#add8e6").pack(padx=10,pady=10,anchor=NW)
metin_alani = Text(frame_sag_ana,width=40,height=10)
metin_alani.tag_configure("style",foreground="#bfbfbf",font=("Verdana",12,"bold"))
metin_alani.pack()

karsilama_metni = "Mesajınızı buraya yazınız..."
metin_alani.insert(END,karsilama_metni,"style")

#4.PART
def gonder():
    from tkinter import messagebox
    son_mesaj = ""
    try:
        if var.get():
            if var.get() == 1:
                son_mesaj += "Veriniz başarı ile kaydedilmiştir."
                tip = hatirlatma_tipi_opsiyon.get() if hatirlatma_tipi_opsiyon.get() == "" else "Genel"
                tarih = hatirlatma_tarih_secici.get()
                mesaj = metin_alani.get("1.0","end")
                with open("hatirlatmalar.txt","w") as dosya:
                    dosya.write(
                        "{} Kategorisinde, {} Tarih ile '{}' notuyla hatırlatmanız oluşturulmuştur".format(
                            tip,
                            tarih,
                            mesaj
                         ))
                    dosya.close
                messagebox.showinfo("Başarılı işlem",son_mesaj)
            elif var.get() == 2:
                son_mesaj += "Veriniz başarı ile kaydedilmiştir."
                messagebox.showinfo("Başarılı işlem",son_mesaj)
        else:
            messagebox.showwarning("Hatalı işlem","Hatalı işlem yaptınız, bir yerleri doldurmayı mı unuttunuz?")
    except:
        messagebox.showerror("Başarısız işlem","İşleminiz başarılı oldu lütfen yaptıklarınızı gözden geçirin.")
    finally:
        master.destroy()
    
    
gonder_buton = Button(frame_sag_ana,text="Gönder",command=gonder)
gonder_buton.pack()


canvas.mainloop()
```
